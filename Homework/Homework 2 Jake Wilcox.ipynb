```python
import numpy as np

from scipy.stats import binom

#inputs to solve
spot_input = eval(input("What is the spot price? "))
strike_input = eval(input("What is the strike price? "))
rate_input = eval(input("What is the interest rate? "))
dividend_input = eval(input("What is the dividend? "))
sigma_input = eval(input("What is sigma? "))
t_input = eval(input("What is T? "))
n_input = eval(input("How many rounds? "))
option_type_input = input("Call or Put? ")
option_origin_input = input("Euro or American? ")
method_input = input("Recursive or risk neutral? ")
     

#Variables based off of inputs 
h = t_input / n_input 
drift = (rate_input - dividend_input)*h
volatility = sigma_input * np.sqrt(h)
u = 1.3
#np.exp(drift + volatility)
d = 0.8
#np.exp(drift - volatility)
pu_star = ((np.exp((rate_input - dividend_input)*h) - d)) / (u - d)
pd_star = (1 - pu_star)


#Euro Option function
def euro_option(spot_input, strike_input, rate_input, n_input, option_type_input, method_input, u, d, pu_star):
    SpotT = np.zeros(n_input+1)
    Probz = np.zeros(n_input + 1)
    Payoffs = np.zeros(n_input + 1)
    discount = np.exp(-rate_input * h)
    for i in range(n_input + 1):
        SpotT[i] = spot_input * (u ** (n_input - i)) * (d ** i)
        Probz[i] = binom.pmf(i, n_input, pd_star)
        if option_type_input == "Call" or option_type_input == "call":
            Payoffs[i] = max(SpotT[i] - strike_input, 0)
        else:
            Payoffs[i] = max(strike_input - SpotT[i], 0)
    print("\n")
    print("The original payoffs are: ", Payoffs)
    if method_input == "risk neutral":
        option_price = '${:,.4f}'.format(np.exp(-(rate_input*t_input)) * np.dot(Payoffs, Probz))
    else:
        for t in range((n_input - 1), -1, -1):
            for j in range(t + 1):
                SpotT[j] = SpotT[j] / u
                Delta = (Payoffs[j] - Payoffs[j+1]) / (SpotT[j] * (u - d))
                Beta = discount * ((u * Payoffs[j+1] - d * Payoffs[j]) / (u - d))
                Payoffs[j] = (SpotT[j] * Delta + Beta)
                option_price = '${:,.4f}'.format(Payoffs[0])
    
    print("\n")
    print("The price of the option is: ", option_price)


#American Option function baby
def american_option(spot_input, strike_input, rate_input, n_input, option_type_input, u, d, pu_star):
    Payoffs = np.zeros(n_input + 1)
    SpotT = np.zeros(n_input + 1)
    discount = np.exp(-rate_input * h)
    for i in range(n_input + 1):
        SpotT[i] = spot_input * (u ** (n_input - i)) * (d ** i)
        if option_type_input == "Call" or option_type_input == "call":
            Payoffs[i] = max(SpotT[i] - strike_input, 0)           
        else:
            Payoffs[i] = max(strike_input - SpotT[i], 0)          
    print("\n")
    print("Original payoffs are: ", Payoffs)
    for t in range((n_input - 1), -1, -1):
        for j in range(t + 1):
            Payoffs[j] = discount * (pu_star * Payoffs[j] + pd_star * Payoffs[j+1])
            SpotT[j] = SpotT[j] / u
            if option_type_input == "Call" or option_type_input == "call":
                Payoffs[j] = max(Payoffs[j], (SpotT[j] - strike_input))         
            else:
                Payoffs[j] = max(Payoffs[j], (strike_input - SpotT[j]))
    option_price = Payoffs[0]
    print("The price of the option is", '${:,.4f}'.format(option_price))
        

#If statements for type of option
if option_origin_input == "Euro" or option_origin_input == "euro":
    euro_option(spot_input, strike_input, rate_input, n_input, option_type_input, method_input, u, d, pu_star)

else:
    american_option(spot_input, strike_input, rate_input, n_input, option_type_input, u, d, pu_star)
```

```python
import numpy as np

from scipy.stats import binom

#inputs to solve
spot_input = eval(input("What is the spot price? "))
strike_input = eval(input("What is the strike price? "))
rate_input = eval(input("What is the interest rate? "))
dividend_input = eval(input("What is the dividend? "))
sigma_input = eval(input("What is sigma? "))
t_input = eval(input("What is T? "))
n_input = eval(input("How many rounds? "))
option_type_input = input("Call or Put? ")
option_origin_input = input("Euro or American? ")
method_input = input("Recursive or risk neutral? ")
     

#Variables based off of inputs 
h = t_input / n_input 
drift = (rate_input - dividend_input)*h
volatility = sigma_input * np.sqrt(h)
u = np.exp(drift + volatility)
d = np.exp(drift - volatility)
pu_star = ((np.exp((rate_input - dividend_input)*h) - d)) / (u - d)
pd_star = (1 - pu_star)


#Euro Option function
def euro_option(spot_input, strike_input, rate_input, n_input, option_type_input, method_input, u, d, pu_star):
    SpotT = np.zeros(n_input+1)
    Probz = np.zeros(n_input + 1)
    Payoffs = np.zeros(n_input + 1)
    discount = np.exp(-rate_input * h)
    for i in range(n_input + 1):
        SpotT[i] = spot_input * (u ** (n_input - i)) * (d ** i)
        Probz[i] = binom.pmf(i, n_input, pd_star)
        if option_type_input == "Call" or option_type_input == "call":
            Payoffs[i] = max(SpotT[i] - strike_input, 0)
        else:
            Payoffs[i] = max(strike_input - SpotT[i], 0)
    print("\n")
    print("The original payoffs are: ", Payoffs)
    if method_input == "risk neutral":
        option_price = '${:,.4f}'.format(np.exp(-(rate_input*t_input)) * np.dot(Payoffs, Probz))
    else:
        for t in range((n_input - 1), -1, -1):
            for j in range(t + 1):
                SpotT[j] = SpotT[j] / u
                Delta = (Payoffs[j] - Payoffs[j+1]) / (SpotT[j] * (u - d))
                Beta = discount * ((u * Payoffs[j+1] - d * Payoffs[j]) / (u - d))
                Payoffs[j] = (SpotT[j] * Delta + Beta)
                option_price = '${:,.4f}'.format(Payoffs[0])
    
    print("\n")
    print("The price of the option is: ", option_price)


#American Option function baby
def american_option(spot_input, strike_input, rate_input, n_input, option_type_input, u, d, pu_star):
    Payoffs = np.zeros(n_input + 1)
    SpotT = np.zeros(n_input + 1)
    discount = np.exp(-rate_input * h)
    for i in range(n_input + 1):
        SpotT[i] = spot_input * (u ** (n_input - i)) * (d ** i)
        if option_type_input == "Call" or option_type_input == "call":
            Payoffs[i] = max(SpotT[i] - strike_input, 0)           
        else:
            Payoffs[i] = max(strike_input - SpotT[i], 0)          
    print("\n")
    print("Original payoffs are: ", Payoffs)
    for t in range((n_input - 1), -1, -1):
        for j in range(t + 1):
            Payoffs[j] = discount * (pu_star * Payoffs[j] + pd_star * Payoffs[j+1])
            SpotT[j] = SpotT[j] / u
            if option_type_input == "Call" or option_type_input == "call":
                Payoffs[j] = max(Payoffs[j], (SpotT[j] - strike_input))         
            else:
                Payoffs[j] = max(Payoffs[j], (strike_input - SpotT[j]))
    option_price = Payoffs[0]
    print("The price of the option is", '${:,.4f}'.format(option_price))
        

#If statements for type of option
if option_origin_input == "Euro" or option_origin_input == "euro":
    euro_option(spot_input, strike_input, rate_input, n_input, option_type_input, method_input, u, d, pu_star)

else:
    american_option(spot_input, strike_input, rate_input, n_input, option_type_input, u, d, pu_star)
```
