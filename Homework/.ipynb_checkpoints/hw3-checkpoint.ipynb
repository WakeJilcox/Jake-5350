# Problem 1.a)

```python
spot = 40.0
strike = 40.0
sigma = 0.30
rate = 0.08
tau = 1
div = 0.0
h = tau / 252
```

```python
import numpy as np
from scipy.stats import norm
   
def bsmCallDelta(spot, strike, sigma, rate, tau, div):
    d1 = (np.log(spot / strike) + (rate - div + 0.5 * sigma * sigma) * tau) / (sigma * np.sqrt(tau))
    delta = np.exp(-div * tau) * norm.cdf(d1)
    
    return delta

def bsmPutDelta(spot, strike, sigma, rate, tau, div):
    d1 = (np.log(spot / strike) + (rate - div + 0.5 * sigma * sigma) * tau) / (sigma * np.sqrt(tau))
    delta = np.exp(-div * tau) * norm.cdf(-d1)
    
    return delta

def bsmCallGamma(spot, strike, sigma, rate, tau, div):
    d1 = (np.log(spot / strike) + (rate - div + 0.5 * sigma * sigma) * tau) / (sigma * np.sqrt(tau))
    gamma = (np.exp(-div * tau) * norm.pdf(1*d1)) / (spot * np.sqrt(tau))
    
    return gamma

def bsmPutGamma(spot, strike, sigma, rate, tau, div):
    d1 = (np.log(spot / strike) + (rate - div + 0.5 * sigma * sigma) * tau) / (sigma * np.sqrt(tau))
    gamma = (np.exp(-div * tau) * norm.pdf(1*d1)) / (spot * np.sqrt(tau))
    
    return gamma

def bsmCallVega(spot, strike, sigma, rate, tau, div):
    d1 = (np.log(spot / strike) + (rate - div + 0.5 * sigma * sigma) * tau) / (sigma * np.sqrt(tau))
    vega = spot * np.exp(-div * tau) * norm.pdf(d1) * np.sqrt(tau)
    
    return vega

def bsmPutVega(spot, strike, sigma, rate, tau, div):
    d1 = (np.log(spot / strike) + (rate - div + 0.5 * sigma * sigma) * tau) / (sigma * np.sqrt(tau))
    vega = spot * np.exp(-div * tau) * norm.pdf(d1) * np.sqrt(tau)
    
    return vega

def bsmCallTheta(spot, strike, sigma, rate, tau, div):
    d1 = (np.log(spot / strike) + (rate - div + 0.5 * sigma * sigma) * tau) / (sigma * np.sqrt(tau))
    d2 = d1 - (sigma * np.sqrt(tau))
    Theta = (div * spot ** (-div * tau)) * norm.cdf(d1) - (rate * strike * np.exp(-rate * tau)) * norm.cdf(d2) - (strike * np.exp(-rate * tau) * norm.pdf(d2) * sigma) / (2 * np.sqrt(tau))
        
    return Theta

def bsmPutTheta(spot, strike, sigma, rate, tau, div):
    d1 = (np.log(spot / strike) + (rate - div + 0.5 * sigma * sigma) * tau) / (sigma * np.sqrt(tau))
    d2 = d1 - (sigma * np.sqrt(tau))
    CallTheta = (div * spot ** (-div * tau)) * norm.cdf(d1) - (rate * strike * np.exp(-rate * tau)) * norm.cdf(d2) - (strike * np.exp(-rate * tau) * norm.pdf(d2) * sigma) / (2 * np.sqrt(tau))
    
    Theta = CallTheta + rate * strike * np.exp(-rate * tau) - div * spot * np.exp(-div * tau)
    return Theta


def bsmCallRho(spot, strike, sigma, rate, tau, div):   
    d1 = (np.log(spot / strike) + (rate - div + 0.5 * sigma * sigma) * tau) / (sigma * np.sqrt(tau))
    d2 = d1 - (sigma * np.sqrt(tau))
    Rho = tau * strike * np.exp(-rate * tau) * norm.cdf(d2)
    
    return Rho

def bsmPutRho(spot, strike, sigma, rate, tau, div):   
    d1 = (np.log(spot / strike) + (rate - div + 0.5 * sigma * sigma) * tau) / (sigma * np.sqrt(tau))
    d2 = d1 - (sigma * np.sqrt(tau))
    Rho = tau * strike * np.exp(-rate * tau) * norm.cdf(-d2)
    
    return Rho
```

# Problem 1. b&c)

```python
import matplotlib.pyplot as plt


call_delta = np.empty(81)
call_gamma = np.empty(81)
call_vega = np.empty(81)
call_theta = np.empty(81)
call_rho = np.empty(81)
put_delta = np.empty(81)
put_gamma = np.empty(81)
put_vega = np.empty(81)
put_theta = np.empty(81)
put_rho = np.empty(81)

for i in range(81):
    call_delta[i] = bsmCallDelta(i, strike, sigma, rate, tau, div)
    call_gamma[i] = bsmCallGamma(i, strike, sigma, rate, tau, div)
    call_vega[i] = bsmCallVega(i, strike, sigma, rate, tau, div)
    call_theta[i] = bsmCallTheta(i, strike, sigma, rate, tau, div)
    call_rho[i] = bsmCallRho(i, strike, sigma, rate, tau, div)
    put_delta[i] = bsmPutDelta(i, strike, sigma, rate, tau, div)
    put_gamma[i] = bsmPutGamma(i, strike, sigma, rate, tau, div)
    put_vega[i] = bsmPutVega(i, strike, sigma, rate, tau, div)
    put_theta[i] = bsmPutTheta(i, strike, sigma, rate, tau, div)
    put_rho[i] = bsmPutRho(i, strike, sigma, rate, tau, div)
    
```

```python
plt.subplot(1,2,1)
plt.plot(call_rho)
plt.title('Rho Call')

plt.subplot(1,2,2)
plt.plot(put_rho)
plt.title('Rho Put')
```
